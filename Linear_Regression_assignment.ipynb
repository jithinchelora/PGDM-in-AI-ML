# Split the Train, Test data

```python
x = df_encoded.drop(['cnt'],axis=1)
```

```python
y = df_encoded['cnt']
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,train_size=0.8,random_state=100)
```

```python
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

# Scaling

```python
x_train
```

```python
#need to do the scaling, trying the standardscalar form sklearn

from sklearn.preprocessing import StandardScaler
standard = StandardScaler()
x_train = standard.fit_transform(x_train)
```

```python
x_test = standard.transform(x_test)
```

```python
x_train
```

```python
x_train_df = pd.DataFrame(x_train,columns=x.columns)
x_test_df = pd.DataFrame(x_test,columns=x.columns)
```

```python
x_train_df
```

# Feature Selection

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression


estimator = LinearRegression()
selector = RFE(estimator,n_features_to_select=5)
selector = selector.fit(x_train_df,y_train)
selector.support_
```

```python
selected_features = x_train_df.columns[selector.support_]
x_train = x_train_df[selected_features]
x_test = x_test_df[selected_features]
```

# Modeling

```python
import statsmodels.api as sm

x_train_sm = sm.add_constant(x_train)
x_test_sm = sm.add_constant(x_test)
model1 = sm.OLS(np.array(y_train),x_train_sm)
res1 = model1.fit()
res1.summary()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant

x_train_vif = add_constant(x_train)

vif_data = pd.DataFrame()
vif_data['Feature'] = x_train_vif.columns
vif_data['VIF']= [variance_inflation_factor(x_train_vif.values,i) for i in range(x_train_vif.shape[1])]
print(vif_data)

```
